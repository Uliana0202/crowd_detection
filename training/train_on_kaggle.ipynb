```python
import os
import yaml
import torch
from ultralytics import YOLO

SEED = 42 
device = 0 if torch.cuda.is_available() else 'cpu'
```

```python
TRAIN_IMAGES = "/kaggle/input/crowd-human-crowd-detection/CrowdHuman dataset/dataset/train/images"
VAL_IMAGES   = "/kaggle/input/crowd-human-crowd-detection/CrowdHuman dataset/dataset/val/images"

data_config = {
    'path': '',
    'train': TRAIN_IMAGES,
    'val': VAL_IMAGES,
    
    'nc': 1,
    'names': {0: 'person'}
}

config_path = '/kaggle/working/crowdhuman_config.yaml'
with open(config_path, 'w') as f:
    yaml.dump(data_config, f)
```

# –î–æ–æ–±—É—á–µ–Ω–∏–µ

```python
model_name = 'yolo11n.pt' if device == 'cpu' else 'yolo11s.pt'
model = YOLO(model_name)
```

```python
results = model.train(
    data=config_path,

    # –í–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç—å
    seed=SEED,
    deterministic=True,
    
    epochs=10,
    patience=3,
    imgsz=640,
    
    batch=16 if device != 'cpu' else 4,
    workers=4 if device != 'cpu' else 0,
    device=device,
    
    pretrained=True,
    optimizer='auto',
    
    # –õ–æ–≥–∏—Ä–æ–≤–∞–Ω–∏–µ
    project='/kaggle/working/runs',
    name='crowd_human_yolo11s',
    exist_ok=True,
    verbose=True
)
```
